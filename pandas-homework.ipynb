# PyCitySchools Homework

```python
# Setup
import pandas as pd
import numpy as np

# Read School and Student Data
school_data = pd.read_csv("D:/pandas-challenge/pandas-challenge/schools_complete.csv")
student_data = pd.read_csv("D:/pandas-challenge/pandas-challenge/students_complete.csv")

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
#finding total number of schools
school_count = school_data_complete['school_name'].nunique()

#counting total students
student_total = school_data_complete['student_name'].count()

#calculating total budget
total_budget = school_data['budget'].sum()

#calculating average math score
average_math = school_data_complete['math_score'].mean()

#calculating average reading score
average_reading = school_data_complete['reading_score'].mean()

#calculating percent of students passing math
percentage_pass_math = school_data_complete['math_score'].where(school_data_complete['math_score']>=70).count() / student_total * 100

#calculating percent of students passing reading
percentage_pass_reading = school_data_complete['reading_score'].where(school_data_complete['reading_score']>=70).count() / student_total * 100

#calculating percent of students passing math AND reading
overall_passing = school_data_complete['student_name'].where((school_data_complete['math_score']>=70) & (school_data_complete['reading_score']>=70)).count() / student_total * 100 
```

```python
#creating dataframe to hold/show all the metrics
dataset1 = {'total schools':school_count,'total students':student_total, 'total budget':total_budget,'average math score':average_math, 'average reading score':average_reading, '%Passing Math':percentage_pass_math, '%passing reading':percentage_pass_reading, '%overall passing':overall_passing}
district_summary = pd.DataFrame(dataset1,index=[0])
district_summary
```

## School Summary

```python
#edited original school_data to be basis for summary
each_school = school_data.drop(columns=['School ID']).set_index('school_name')
```

```python
#adding per student budget column
each_school['per student budget'] = each_school['budget']/each_school['size']
```

```python
#getting average math and reading scores by school
average_per_school = school_data_complete.drop(columns=['School ID','Student ID','size','budget']).groupby(by = 'school_name').mean()
```

```python
#percent passing math by school
passing_grades_math = school_data_complete.loc[school_data_complete['math_score'] >= 70,:]
grouped_math = passing_grades_math.groupby(by = 'school_name').count()
percent_math = grouped_math['math_score']/each_school['size'] * 100
```

```python
#percent passing reading
passing_grades_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70,:]
grouped_reading = passing_grades_reading.groupby(by = 'school_name').count()
percent_reading = grouped_reading['reading_score']/each_school['size'] * 100
```

```python
#percent passing overall
overall_passes = school_data_complete.loc[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70),:]
grouped_overall = overall_passes.groupby(by = 'school_name').count()
percent_overall = grouped_overall['reading_score']/each_school['size'] * 100
```

```python
#merging two datasets
school_partial_sum = pd.merge(each_school,average_per_school, how = 'left',on = 'school_name') 
```

```python
#adding the rest of the columns to the dataset
school_summary = pd.concat([school_partial_sum,percent_reading,percent_math,percent_overall], axis=1)
```

```python
#permanently sorting dataset by school name
school_summary.sort_index(inplace=True)
```

```python
#renaming columns for final output
school_summary_final = school_summary.rename(columns = {'type':'school type','size':'total students','budget':'total school budget','reading_score':'average reading score','math_score':'average math score',0:'% passing reading',1:'% passing math',2:'% overall passing'})
```

## Scores by School Spending

```python
#creating conditions for new column
conditions = [
    (school_summary_final['per student budget'] < 585),
    (school_summary_final['per student budget'] >=585) & (school_summary_final['per student budget'] < 630),
     (school_summary_final['per student budget'] >=630) & (school_summary_final['per student budget'] < 645),
     (school_summary_final['per student budget'] >=645) & (school_summary_final['per student budget'] < 680)
]

#creating values for new column
values = ['<$585','$585-630','$630-645','$645-680']

#creating new column for spending ranges
school_summary_final['Spending Ranges (Per Student)'] = np.select(conditions, values)
```

```python
#averaging grouped by spending bins created above
spending = school_summary_final.drop(columns = ['total students','total school budget','per student budget']).groupby(by = 'Spending Ranges (Per Student)').mean()
```

## Scores by School Size

```python
#creating conditions for new column
conditions2 = [
    (school_summary_final['total students'] < 1000),
    (school_summary_final['total students'] >=1000) & (school_summary_final['total students'] < 2000),
     (school_summary_final['total students'] >=2000) & (school_summary_final['total students'] < 5000)
]

#creating values for new column
values2 = ['Small (<1000)','Medium (1000-2000)','Large (2000-5000)']

#creating new column for school size bins
school_summary_final['School Size'] = np.select(conditions2, values2)
```

```python
#calculating averages grouped by school size
school_size = school_summary_final.drop(columns = ['total students','total school budget','per student budget']).groupby(by = 'School Size').mean()
```
